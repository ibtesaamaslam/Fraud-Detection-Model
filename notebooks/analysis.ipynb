```python
# ---------------------------------------------------------------------------
# Cell 1 — Imports and Setup
# ---------------------------------------------------------------------------

from src.utils.helpers import ensure_project_directories, set_seed
from src.utils.config import RANDOM_STATE
from src.data.load_data import load_data
from src.data.preprocess import preprocess_data
from src.models.train_model import train_models
from src.models.evaluate_model import evaluate_models
from src.visualization.eda import run_eda

ensure_project_directories()
set_seed(RANDOM_STATE)
```

```python
# ---------------------------------------------------------------------------
# Cell 2 — Load Raw Data
# ---------------------------------------------------------------------------

df = load_data()
df.head()
```

```python
# ---------------------------------------------------------------------------
# Cell 3 — Exploratory Data Analysis (EDA)
# ---------------------------------------------------------------------------

run_eda(df)
```

```python
# ---------------------------------------------------------------------------
# Cell 4 — Preprocess the Data
# ---------------------------------------------------------------------------

X_train, X_test, y_train, y_test, scaler, feature_names = preprocess_data(df)
X_train.head()
```

```python
# ---------------------------------------------------------------------------
# Cell 5 — Train All Candidate Models
# ---------------------------------------------------------------------------

models, X_res, y_res = train_models(X_train, y_train)
```

```python
# ---------------------------------------------------------------------------
# Cell 6 — Evaluate and Compare All Models
# ---------------------------------------------------------------------------

metrics_df = evaluate_models(models, X_test, y_test)
metrics_df
```

```python
# ---------------------------------------------------------------------------
# Cell 7 — Train the Best Model on the Full Training Data and Save It
# ---------------------------------------------------------------------------

from src.explainability.shap_explainer import save_shap_summary_plot
best_model = models[metrics_df.iloc[0]["model"]]
sample = X_test.sample(min(300, len(X_test)), random_state=RANDOM_STATE)
save_shap_summary_plot(best_model, sample)
```
